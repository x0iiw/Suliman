## ETS Decomposition of Airline Passengers Dataset:

## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

## Read dataset

```python
df = pd.read_csv('/content/airline-passengers.csv',index_col ='Month',
                       parse_dates = True)
```

## show dataset

```python
df.head()
```

## Apply ETS

```python
result = seasonal_decompose(df['Passengers'],model='multiplicative')
```

## Plot ETS

```python
#ETS ploting
result.plot()
```
