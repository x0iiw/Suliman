## ETS Decomposition of Airline Passengers Dataset:

## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

## Read dataset

```python
df = pd.read_csv('/content/traffic.csv',index_col ='DateTime',
                       parse_dates = True)
```

## show dataset

```python
df.head()
```

```python
df.shape
```

## Apply ETS

```python
result = seasonal_decompose(df['Vehicles'],model='multiplicative',period=30)
```

## Plot ETS

```python
#ETS ploting

result.plot()
plt.xticks(rotation=90)
plt.show()
```
